```python
def generate_images(validation, ground_truth, prediction):
  plt.figure(figsize=(15,15))

  display_list = [validation[0], ground_truth[0], prediction[0]]
  title = ['Input Image', 'Ground Truth', 'Predicted Image']

  for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.title(title[i])
    # getting the pixel values between [0, 1] to plot it.
    plt.imshow(display_list[i])
    plt.axis('off')
  plt.show()
```

```python
import matplotlib.image as mpimg
import os , glob
import matplotlib.pyplot as plt

images = os.listdir("color/")
truths = os.listdir("ground/")
validate = os.listdir("validation/")
degrades = os.listdir("color2/")


test_images = []
ground_truth = []
validation = []
degradation = []
display_list = []

for img in images:
  test_images.append(mpimg.imread("color/" + img))
for img in truths:
  ground_truth.append(mpimg.imread("ground/" + img))
for img in validate:
  validation.append(mpimg.imread("validation/" + img))
for img in degrades:
  degradation.append(mpimg.imread("color2/" + img))

for i in range (10):
  display_list.append(validation[i])
  display_list.append(ground_truth[i])
  display_list.append(test_images[i])
  display_list.append(degradation[i])

title = ['Input Image', 'Ground Truth', 'Predicted Image']
plt.figure(figsize=(15,15))
 
for i in range(40):
  plt.subplot(10, 4, i+1)
  #plt.title(title[i])
  plt.imshow(display_list[i])
plt.show()
#generate_images(validation, ground_truth, test_images)
  #plt.figure()
  #plt.imshow(mpimg.imread("color/" + img))
  #plt.imshow(mpimg.imread("ground_truth/" + img))
```

```python
import cv2
import numpy
import matplotlib.pyplot as plt
predict = cv2.imread("./color4/b3599e04-174421e4.png")
ground = cv2.imread("./ground2/b3599e04-174421e4.png")
#plt.imshow(cv2.cvtColor(predict, cv2.COLOR_BGR2RGB))
#plt.imshow(cv2.cvtColor(ground, cv2.COLOR_BGR2RGB))
# IoU calculation
intersection = numpy.logical_and(predict, ground)
union = numpy.logical_or(predict, ground)
iou_score = numpy.sum(intersection) / numpy.sum(union)
print("IoU is %s" % iou_score)
```

```python
# An example of first bounding box
img = cv2.imread("./color2/b21547c1-73e457f8.png")
first_bb_points = [[0, 0], [0, 337], [600, 0], [600, 337]]
stencil = numpy.zeros(img.shape).astype(img.dtype)
contours = [numpy.array(first_bb_points)]
color = [255, 255, 255]
cv2.fillPoly(stencil, contours, color)
result1 = cv2.bitwise_and(img, stencil)
result1 = cv2.cvtColor(result1, cv2.COLOR_BGR2RGB)
plt.imshow(result1)
```
